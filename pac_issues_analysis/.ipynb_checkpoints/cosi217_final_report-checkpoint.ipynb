### Topic Annotation with Snorkel
This project applies snorkel label functions and uses supervised learning to classify the labels for the issues. For annotation, there are 13~14 major categories. Each label can be assigned in terms of the keyword or if word_n appeared in the text.

    “Snorkel is a system for programmatically building and managing training datasets without manual labeling. In Snorkel, users can develop large training datasets in hours or days rather than hand-labeling them over weeks or months.”
    

For more info:
- [Using Snorkel For Multi-Label Annotation](https://towardsdatascience.com/using-snorkel-for-multi-label-annotation-cc2aa217986a)

label preprocessing:

```python
def topic_labeling_preprocessing(self):
    topics = self.feat_dict['keywords']['topic'].values.tolist()
    label_dict = {topics[k]:k for k in range(len(topics))}

    for item in self.feat_dict.keys():
      if item == 'required':
        flag = True
      self.feat_dict[item] = BLTopicModelingHelperFunctions().clean_keywords(self.feat_dict[item], topics, required_word=True)
    self.feat_dict['labels'] = label_dict

    lfs = []
    for i in range(len(topics)):
      print(f'topic {i+1}:     ', topics[i])
      print(f'keyword {i+1}:   ', self.feat_dict['keywords'][topics[i]])
      print(f'not_match {i+1}: ', self.feat_dict['not_match'][topics[i]])
      print(f'required {i+1}:  ', self.feat_dict['required'][topics[i]])
      print(f'label {i+1}:     ', self.feat_dict['labels'][topics[i]])
      lfs.append(self.topic_match_lf([topics[i]], topics, i))
    return topics, lfs
```

label training:

```python

class SnorkelTopicTraining:

  def __init__(self, input_path, topics, lfs):
    self.label_model = LabelModel(cardinality=len(topics), verbose=True)
    self.df = Preprocessing().create_explode_df(pd.read_csv(input_path, encoding='utf-8'),'text')
    self.lfs = lfs



  def train(self, how_many_rows_to_train=None, output_name = 'tss.csv'):
    df_train = self.df.copy()

    #create lfs for each topic class
    applier = PandasLFApplier(lfs=self.lfs)
    L_train = applier.apply(df=df_train)
    print('Start topic training')
    self.label_model.fit(L_train, n_epochs=500, log_freq=50, seed=123) # original n_epochs: 500 but it was changed because of fast execution
    df_train["class_simple"] = self.label_model.predict(L=L_train, tie_break_policy="abstain")
    df_train = df_train[df_train["class_simple"] != '-2']
    df_train['topics'] = df_train['class_simple'].apply(str).apply(BLTopicModelingHelperFunctions(topics=topics).assign_topics)
    df_train = df_train[df_train['topics']!= 'NOT_RELEVANT']
    df_train.to_csv(output_name) # Save training dataset for fasttext topic labeling
#     print(df_train)
    return df_train
```

After classification, each label is manually checked for accuracy on small, medium, and large validation sets, while a group of manually defined keywords of the label are constantly adjusted according to the accuracy of the label training results.

## Training and testing with fasttext 

Finally, labeled text are separated into two .csv files one for topic and the other for sentiment analysis. This projects adopts fasttext for topic classification and sentiment training.

    "fastText is a library for learning of word embeddings and text classification created by Facebook's AI Research lab. The model allows one to create an unsupervised learning or supervised learning algorithm for obtaining vector representations for words. Facebook makes available pretrained models for 294 languages."

```python
class FastTopicSentTraining:

  def __init__(self,snorkel_training_df, sentiment_training_csv,df):
    self.sent_training_df = sentiment_training_csv
    self.topic_train_df = snorkel_training_df
    self.df = df

  # topic modeling ('class_simple' is the name of topic labels column in training df)
  def training(self):
    topic_df = fast_process_io(snorkel_training_df, df, 'class_simple', ext_path,
                               0.1,  # lr_in (learn rate)
                               70,  # epoch_in
                               2,  # ngrams_in
                               1000,  # size_each
                               0.5,  # threshold_in  threshold to filter our low accuracy
                               'multi',  # single_multi
                               0,     # minimum ngrams
                               0)    # maximum ngrams

    topic_df['class_simple_pred_label'] = topic_df['class_simple_pred_label'].fillna(
      '')
    topic_df['class_simple_pred_label'] = topic_df['class_simple_pred_label'].apply(
      lambda x: x.replace('__label__', ''))
    topic_df['class_simple_pred_label2'] = topic_df['class_simple_pred_label2'].fillna(
      '')
    topic_df['class_simple_pred_label2'] = topic_df['class_simple_pred_label2'].apply(
      lambda x: x.replace('__label__', ''))

    #sentiment predict ('sent_score' is the name of sentiment good/bad column in training df)

    sent_df = fast_process_io(self.sent_training_df, df, 'sent_score', ext_path,
                              .1,  # lr_in
                              70,  # epoch_in
                              2,  # ngrams_in
                              800,  # size_each
                              0.5,  # threshold_in
                              'single',  # single_multi
                              0,        # minimum ngrams
                              0)        # maximum ngrams


    sent_df = sent_df[['extra_id', 'sent_score_pred_label', 'sent_score_pred_acc']]

    sent_df['sent_score_pred_label'] = sent_df['sent_score_pred_label'].fillna('')
    sent_df['sent_score_pred_label'] = sent_df['sent_score_pred_label'].apply(
      lambda x: x.replace('__label__', ''))

    # clean up sentiment
    sent_df['sent_score_pred_label'] = sent_df['sent_score_pred_label'].replace(
      'good', 1)
    sent_df['sent_score_pred_label'] = sent_df['sent_score_pred_label'].replace(
      'bad', -1)
    sent_df['sent_score_pred_label'] = sent_df['sent_score_pred_label'].replace(
      'neutral', 0)
    return topic_df, sent_df
```

- training and testing. I used a simple logistic regression model and calculated the accuracy along precison and recall score with fasttext.

```python
def fast_process_io(train_csv_df_path, to_run_df, label, ext_path, lr_in,
                    epoch_in, ngrams_in, size_each, threshold_in, single_multi, minn, maxn):

  train_csv_df = pd.read_csv(train_csv_df_path)
  train_csv_df = train_csv_df.rename(columns={label: "label"})
  train_csv_df['orig_label'] = train_csv_df['label'].copy()
  train_csv_df = train_csv_df[pd.notnull(train_csv_df['label'])] # drop not classified tweets
  try:
    train_csv_df = train_csv_df[train_csv_df['label'] != 'no_topic']
  except:
    pass


  train_csv_df = train_csv_df[['label', 'text']] # remove any extra columns
  print('training dataset prep')
  train_csv_df['label'] = train_csv_df['label'].astype(str)
  train_csv_df = prep_df_labeling(train_csv_df)

  # check distribution
  print(train_csv_df.groupby('label').count())
  train_csv_df = train_csv_df.groupby('label').filter(lambda x: len(x) >= 10)
  size_ls = list(train_csv_df['label'].value_counts())



  # make equal number of each label
  size = np.percentile(size_ls, 80)      # sample size percentile
  size = int(round(size))
  print("The avg size for each topic:")
  print(size)
  print(train_csv_df)

  replace = True  # with replacement (if you want it to select same row again)
  fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
  train_csv_df = train_csv_df.groupby('label', as_index=False).apply(fn)

  train_csv_df['label_text'] = train_csv_df['label'] + ' ' + train_csv_df['text']
  train_csv_df = train_csv_df['label_text']


  # reset index before exporting
  train_csv_df = train_csv_df.reset_index(drop=True)

  # shuffle order ()
  train_csv_df = train_csv_df.sample(frac=1, random_state=123).reset_index(drop=True)

  train_csv_df.iloc[0:int(len(train_csv_df)*0.4)].to_csv(ext_path +
                                                         str() + '_train.txt',
                                                         index=False, header=False,
                                                         sep="\t", quoting=csv.QUOTE_NONE,
                                                         quotechar="",  escapechar="\\")

  train_csv_df.iloc[int(len(train_csv_df)*0.4):int(len(train_csv_df)*0.6)].to_csv(ext_path +
                                                                                  str() + '_valid.txt',
                                                                                  index=False, header=False,
                                                                                  sep="\t", quoting=csv.QUOTE_NONE,
                                                                                  quotechar="",  escapechar="\\")

  train_csv_df.iloc[int(len(train_csv_df)*0.6):].to_csv(ext_path +
                                                        str() + '_test.txt',
                                                        index=False, header=False,
                                                        sep="\t", quoting=csv.QUOTE_NONE,
                                                        quotechar="",  escapechar="\\")

  train_in = ext_path + str() + '_train.txt'
  valid_in = ext_path + str() + '_valid.txt'
  test_in = ext_path + str() + '_test.txt'

  print('exported training and valid sets')
  if single_multi == 'single':
    model = fasttext.train_supervised(input=train_in,
                                      lr=lr_in, epoch=epoch_in,
                                      wordNgrams=ngrams_in,
                                      minn= minn,
                                      maxn = maxn)
  elif single_multi == 'multi':
    model = fasttext.train_supervised(input=train_in,
                                      lr=lr_in, epoch=epoch_in,
                                      wordNgrams=ngrams_in,
                                      dim=50, loss='ova',
                                      minn=minn,
                                      maxn= maxn)

  # print()
  # print('train df accuracy is...')
  # print(model.test(train_in))
  # print()
  print('valid data accuracy is...')
  print(model.test(valid_in))
  print()
  print('test data accuracy is...')
  print(model.test(test_in))
  print()
  print('prepare entire dataset')

  to_run_df = to_run_df[pd.notnull(to_run_df['text'])]
  to_run_df = to_run_df[to_run_df['text'] != '']

  to_run_df['text'] = to_run_df['text'].copy()
  to_run_df = to_run_df[pd.notnull(to_run_df['text'])]
```

### Discussion  

This project:

- explored both automated and manual annotation
- explored different annotation tools/techniques and machine learning packages
- explored natural language annotation and processing in a more data-science like way

In the process, encountered:

- difficulty generating fine-grained granular data 

- issues with large web-scraped data for clean-up

- certain limitations go along with the choice of an annotation tool/technique or a form of annotation
    - manual: time-consuming and issues with settling on a set of labels
    - auto: requires novel ways to adjudicate since data can run super large
- problems with overfitting 


Future prospect

- since this project is done single-handly by myself, unfortunately there was no chance that the IAA (the Cohen Kappa or the Fleiss Kappa) was calculated. In the future, more scrunity in the adjudication process.

- try different classification model; including the deeper model(CRF, BiLSTM, BERT,Transformer,etc)
